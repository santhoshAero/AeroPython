```python
import math
import numpy as np
from matplotlib import pyplot
N=50
xstart,xend=-2,2
ystart,yend=-2,2
x=np.linspace(xstart,xend,N)
y=np.linspace(ystart,yend,N)

X,Y=np.meshgrid(x,y)
pyplot.scatter(X,Y,s=5)
```

```python
# Make source and sink
s1_str=-5
x_1,y_1=-1,0
u_s1=s1_str*(X-x_1)/((X-x_1)**2+(Y-y_1)**2)
v_s1=s1_str*(Y-y_1)/((X-x_1)**2+(Y-y_1)**2)
pyplot.streamplot(X,Y,u_s1,v_s1,density=3, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_1, y_1,color='#CD2305', s=80, marker='o');
phi1=-1*s1_str/2*np.log((X-x_1)**2+(Y-y_1)**2)
pyplot.contour(X,Y,phi1,10);
```

```python
s2_str=-5
x_2,y_2=1,0
u_s2=s2_str*(X-x_2)/((X-x_2)**2+(Y-y_2)**2)
v_s2=s2_str*(Y-y_2)/((X-x_2)**2+(Y-y_2)**2)
phi2=-1*s2_str/2*np.log((X-x_2)**2+(Y-y_2)**2)

s3_str=1
x_3,y_3=0,1
u_s3=s3_str*(X-x_3)/((X-x_3)**2+(Y-y_3)**2)
v_s3=s3_str*(Y-y_3)/((X-x_3)**2+(Y-y_3)**2)
phi3=-1*s3_str/2*np.log((X-x_3)**2+(Y-y_3)**2)

s4_str=1
x_4,y_4=0,-1
u_s4=s4_str*(X-x_4)/((X-x_4)**2+(Y-y_4)**2)
v_s4=s4_str*(Y-y_4)/((X-x_4)**2+(Y-y_4)**2)
phi4=-1*s4_str/2*np.log((X-x_4)**2+(Y-y_4)**2)

u_net=u_s1+u_s2+u_s3+u_s4
v_net=v_s1+v_s2+v_s3+v_s4
phi_net=phi1+phi2+phi3+phi4
pyplot.streamplot(X,Y,u_net,v_net,density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_1, y_1,color='#CD2305', s=80, marker='o');
pyplot.scatter(x_2, y_2,color='#CD2305', s=80, marker='o');
pyplot.scatter(x_3, y_3,color='#CD2305', s=80, marker='o');
pyplot.scatter(x_4, y_4,color='#CD2305', s=80, marker='o');
pyplot.contour(X,Y,phi_net,10);


```
